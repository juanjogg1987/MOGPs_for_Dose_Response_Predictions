```python
# add column for fd_num -GDSC2
fd_num_col = []
for dose in scaled_normalised_GDSC2['dose']:
    if dose =='D7':
        fd_num_col.append('fd_num_1')
    elif dose == 'D6':
        fd_num_col.append('fd_num_2')
    elif dose == 'D5':
        fd_num_col.append('fd_num_3')
    elif dose == 'D4':
        fd_num_col.append('fd_num_4')
    elif dose == 'D3':
        fd_num_col.append('fd_num_5')
    elif dose == 'D2':
        fd_num_col.append('fd_num_6')
    elif dose == 'D1':
        fd_num_col.append('fd_num_7')
    else:
        fd_num_col.append(np.nan)   
scaled_normalised_GDSC2['fd_num_col']= fd_num_col

scaled_normalised_GDSC2
```

```python
# create key for duplicates
scaled_normalised_GDSC2['KEY'] = scaled_normalised_GDSC2.groupby(['SCAN_DRUG_COSMICID','fd_num_col']).cumcount()
scaled_normalised_GDSC2
```

```python
scaled_normalised_GDSC2['SCAN_DRUG_COSMICID_KEY'] = scaled_normalised_GDSC2['SCAN_DRUG_COSMICID'].astype(str)+"_"+ scaled_normalised_GDSC2['KEY'].astype(str)
scaled_normalised_GDSC2
```

```python
# transpose normalised_intensity column
GDSC2_dose_response = scaled_normalised_GDSC2.pivot_table(index=['COSMIC_ID','DRUG_ID','SCAN_DRUG_COSMICID','SCAN_DRUG_COSMICID_KEY','MAX_CONC','FOLD_RANGE'], columns='fd_num_col' , values=['fd_num', 'normalized_intensity']).reset_index()
GDSC2_dose_response


```

```python
GDSC2_dose_response.columns = list(map("_".join, GDSC2_dose_response.columns))
GDSC2_dose_response
```

```python
GDSC2_dose_response.rename (columns= {'COSMIC_ID_':'COSMIC_ID', 
                                      'DRUG_ID_':'DRUG_ID', 
                                      'SCAN_DRUG_COSMICID_':'SCAN_DRUG_COSMICID', 
                                      'SCAN_DRUG_COSMICID_KEY_':'SCAN_DRUG_COSMICID_KEY', 
                                      'MAX_CONC_':'MAX_CONC', 
                                      'FOLD_RANGE_':'FOLD_RANGE'}, inplace=True)
```

```python
GDSC2_dose_response.columns = GDSC2_dose_response.columns.str.replace('fd_num_fd_num_', 'fd_num_')
GDSC2_dose_response.columns = GDSC2_dose_response.columns.str.replace('normalized_intensity_fd_num_', 'norm_cells_')
GDSC2_dose_response
```

```python
GDSC2_dose_response['fd_num_0'] = 0
GDSC2_dose_response['norm_cells_0'] = 1.0
GDSC2_dose_response
```

```python
column_names = ['COSMIC_ID', 'DRUG_ID', 'SCAN_DRUG_COSMICID', 'SCAN_DRUG_COSMICID_KEY',
                'MAX_CONC', 'FOLD_RANGE', 'fd_num_0', 'fd_num_1', 'fd_num_2', 'fd_num_3',
                'fd_num_4', 'fd_num_5', 'fd_num_6', 'fd_num_7','norm_cells_0','norm_cells_1',
                'norm_cells_2', 'norm_cells_3', 'norm_cells_4', 'norm_cells_5',
                'norm_cells_6', 'norm_cells_7']
GDSC2_dose_response = GDSC2_dose_response.reindex(columns=column_names)
GDSC2_dose_response
```
