```python
#GDSC1
GDSC1_drug_response = pd.read_csv(directory_path + "GDSC1_processed_dose_response.csv")
GDSC1_drug_response
```

Merging cell line features - tissue types

```python
# read cell line details downloaded from GDSC database
cell_line_features = pd.read_csv(directory_path + "GDSC_Cell_Lines_Details_Oct2022.csv")
cell_line_features
```

```python
# Retrieve cosmic ID cancer type features and rename a column
features_for_merging = cell_line_features.loc[:, ['COSMIC_ID', 'GDSC_Tissuedescriptor2', 'Cancer Type\n(matching TCGA label)']].rename(columns={'Cancer Type\n(matching TCGA label)': 'Cancer_type_TCGA'})
features_for_merging
```

GDSC1

```python
# merge GDSC1 drug curves with cancer subtype features
GDSC1_drug_response_2fold_tissue = GDSC1_2fold.merge(features_for_merging, on='COSMIC_ID', how='left', indicator=True)
GDSC1_drug_response_4fold_tissue = GDSC1_4fold.merge(features_for_merging, on='COSMIC_ID', how='left', indicator=True)
```

```python
# check -left only
check_GDSC1_drug_response_2fold_tissue = GDSC1_drug_response_2fold_tissue[GDSC1_drug_response_2fold_tissue['_merge'] == 'left_only']
check_GDSC1_drug_response_2fold_tissue
```

```python
# check -left only
check_GDSC1_drug_response_4fold_tissue = GDSC1_drug_response_4fold_tissue[GDSC1_drug_response_4fold_tissue['_merge'] == 'left_only']
check_GDSC1_drug_response_4fold_tissue
```

```python
#drop merge column
GDSC1_drug_response_2fold_tissue_filt = GDSC1_drug_response_2fold_tissue.drop(['_merge'], axis = 1)
GDSC1_drug_response_4fold_tissue_filt = GDSC1_drug_response_4fold_tissue.drop(['_merge'], axis = 1)
GDSC1_drug_response_2fold_tissue_filt
```
